```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing 365 days in a year
df_date = pd.read_excel("DateInOneYear.xlsx",sheetname='2016', header=None)
RevisedDate = df_date[0]
RevisedDate.shape
```

```python
df = pd.read_excel("04. elspot-volumes_2016_daily.xlsx", header = 2, parse_dates=True)
df['Date_daily'] = RevisedDate
df.head(5)
```

```python
df.shape
```

```python
X = RevisedDate
SE1_Buy = df['SE1 Buy']
SE1_Sell = df['SE1 Sell']
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 10), dpi=80, facecolor='w', edgecolor='k')

plt.xlabel('2016-Day')
plt.ylabel('Amount')
plt.title('2016 Sell and Buy')


plt.plot(X,SE1_Buy)
plt.plot(X,SE1_Sell)

plt.legend(loc='upper left')
```
