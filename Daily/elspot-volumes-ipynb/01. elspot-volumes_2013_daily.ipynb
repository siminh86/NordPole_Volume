```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(20, 10)})

import os
```

```python
df = pd.read_csv('/home/paperspace/NordPole_Volume/Daily/CSVFiles/01. elspot-volumes_2013_daily.csv')
```

```python
df.shape
```

```python
df['DateAndTime'] = pd.date_range('1/1/2013', periods=365, freq='D')
```

```python
SE1_Buy = df['SE1 Buy']
SE1_Sell = df['SE1 Sell']
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 10), dpi=80, facecolor='w', edgecolor='k')

plt.plot(df.DateAndTime,SE1_Buy)
plt.plot(df.DateAndTime,SE1_Sell)

plt.xlabel('Year 2013, Daily')
plt.ylabel('Price')
plt.title('Sweden section 1 sell and buy price')

```
