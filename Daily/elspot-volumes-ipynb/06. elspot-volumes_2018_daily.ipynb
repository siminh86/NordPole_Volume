```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/simhaj-local/Documents/GitHub/NordPool/Daily/CSVFiles/06. elspot-volumes_2018_daily.csv')
```

```python
df.shape
```

```python
df['DateAndTime'] = pd.date_range('1/1/2018', periods=365, freq='D')
```

```python
SE1_Buy = df['SE1 Buy']
SE1_Sell = df['SE1 Sell']
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 10), dpi=80, facecolor='w', edgecolor='k')

plt.xlabel('2018-Day')
plt.ylabel('Amount')
plt.title('2018 Sell and Buy')


plt.plot(df.DateAndTime,SE1_Buy)
plt.plot(df.DateAndTime,SE1_Sell)

plt.legend(loc='upper left')
```
