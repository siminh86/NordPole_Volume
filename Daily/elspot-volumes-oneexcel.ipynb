```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing 365 days in a year
df_date = pd.read_excel("elspot-volumes-oneexcel.xlsx",header=None)
RevisedDate = df_date[0]
RevisedDate.shape
```

```python
df = pd.read_excel("elspot-volumes-oneexcel.xlsx")
df['Date_daily'] = RevisedDate
```

```python
df.shape
```

```python
X = df['Date_daily']
SE1_Buy = df['SE1 Buy']
SE1_Sell = df['SE1 Sell']
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 10), dpi=80, facecolor='w', edgecolor='k')

plt.plot(X,SE1_Buy)
plt.plot(X,SE1_Sell)
```
